# 1. List (Creation, Modification and Access):

```python
# a. Create a list named age_list with five integer elements. For eg., [24, 25, 26, 27, 28]
age_list = [24, 25, 26, 27, 28]
print(f"age_list: {age_list}")
```

```python
# b. Create a list named name_list with five string elements.
name_list = ["Arun", "Bala", "Cathy", "Deepa", "Elango"]
print(f"name_list: {name_list}")
```

# 2 List Operations / Modifications:

```python
# a. Append the string "Yazhini" to name_list.
name_list.append("Yazhini")
print(f"name_list after append: {name_list}")
```

```python
# b. Insert the element 30 at index 2 in age_list.
age_list.insert(2, 30)
print(f"age_list after insert: {age_list}")
```

```python
# c. Remove the string "Yazhini" from name_list.
name_list.remove("Yazhini")
print(f"name_list after remove: {name_list}")
```

```python
# d. Pop the last element from age_list.
popped_age = age_list.pop()
print(f"age_list after pop: {age_list}. Popped element: {popped_age}")
```

```python
# e. Extend the age_list with additional ages [29, 30, 26].
age_list.extend([29, 30, 26])
print(f"age_list after extend: {age_list}")
```

```python
# f. Sort age_list in descending order.
age_list.sort(reverse=True)
print(f"age_list after descending sort: {age_list}")
```

```python
# g. Find Max age, Min age and sum of all ages from age_list.
max_age = max(age_list)
min_age = min(age_list)
sum_ages = sum(age_list)
print(f"Max age: {max_age}, Min age: {min_age}, Sum of ages: {sum_ages}")
```

# Dictionary (Creation, Modification and Access):


```python
# a. Create a dictionary named student_marks that maps the names of five students to their marks (use scale of from 0 to 100).
student_marks = {
    "Priya": 92,
    "Kavin": 78,
    "Santhosh": 85,
    "Muthu": 67,
    "Nisha": 95
}
print(f"student_marks dictionary: {student_marks}")
```

```python
# b. Access and print the mark of a specific student, of your choice.
student_name = "Priya"
priya_mark = student_marks[student_name]
print(f"The mark of {student_name} is: {priya_mark}")
```

```python
# c. Add a new student "Janani" with a mark of 80 to the student_marks dictionary.
student_marks["Janani"] = 80
print(f"student_marks after adding Janani: {student_marks}")
```

```python
# d. Update the mark of any one older student to 82.
student_marks["Kavin"] = 82
print(f"student_marks after updating Kavin's mark: {student_marks}")
```

```python
# e. Use the keys(), values(), and items() methods to print all keys, values, and key-value pairs in the student_marks dictionary.
print(f"Keys (Student Names): {student_marks.keys()}")
print(f"Values (Marks): {student_marks.values()}")
print(f"Items (Key-Value Pairs): {student_marks.items()}")
```

# Sets (Operations):

```python
# a. Create a set called my_set with following values: ['a','e','i','o','u', 'a', 'a', 'i']
my_set = set(['a','e','i','o','u', 'a', 'a', 'i'])
print(f"my_set: {my_set}")
# Analyse the output and provide explanation for the same.
print("Explanation: A set stores only unique elements. Duplicate values ('a' and 'i') are automatically discarded during creation.")
```

```python
# b. Attempt to change the value of my_set[4] = 's'. If code throws an error, provide an explanation.
my_set[4] = 's'
```

```python
print(f"Attempt to change my_set[4] = 's' resulted in an Error:'set' object does not support item assignment")
```

```python
# c. Create two sets: set1 with values: {1, 3, 5, 7, 9} and set2 with values: {2, 3, 5, 8, 10}
set1 = {1, 3, 5, 7, 9}
set2 = {2, 3, 5, 8, 10}
print(f"set1: {set1}")
print(f"set2: {set2}")
```

```python
# d. Compute and print the union and intersection of set1 and set2.
union_set = set1.union(set2)
intersection_set = set1.intersection(set2)
print(f"Union of set1 and set2: {union_set}")
print(f"Intersection of set1 and set2: {intersection_set}")
```

# Operators & Conditional Statements : (IF, ELIF, ELSE)

```python
# 1. Prompt user for Input. Score range should be from 0 to 10 (both inclusive).
# Get user input and convert it to an integer
score = int(input("Enter your score (0 to 10): "))

# Input validation (Optional but good practice)
if not (0 <= score <= 10):
    print("Error: Score must be between 0 and 10 (inclusive).")
```

```python
# 2. Find the performance category based on the input score
    # a. Above Average: Score greater than 7
if score > 7:
    category = "Above Average"
    prompt = "Excellent work! Keep setting higher goals."

    # b. Average: Score between 4 and 7 (both inclusive)
elif 4 <= score <= 7:
    category = "Average"
    prompt = "Good effort! Keep practicing, there's room for improvement."

    # C. Below Average: Score lesser than 4
else: # This covers scores less than 4 (0, 1, 2, 3)
    category = "Below Average"
        # Example prompt for Below Average
    prompt = "Need to Improve your performance, consistent practice will lead to better results."
```

```python
# 3. Output: Print the Performance category
# This print statement executes only if the score was valid (0-10)
if 'category' in locals():
    print(f"{category}: {prompt}")
```

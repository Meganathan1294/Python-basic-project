## Prompt the user

```python
number = random.randint(1, 10)

attempts = 3

while attempts > 0:
    guess = int(input("Guess the number (1 to 10): "))

    if guess == number:
        print("Correct! You guessed the number.")
        break
    else:
        attempts -= 1
        print("Wrong guess!")

    if attempts > 0:
        print("Attempts left:", attempts)
    else:
        print("Game Over! The correct number was", number)
```

```python
import random


secret_number = random.randint(1,10)
attempts = 3

while attempts > 0:
    try:
        guess = int(input("Guess the number (1 to 10): "))
    except ValueError:
        print("Invalid input! Please enter a whole number.")
        continue

    if guess < 1 or guess > 10:
        print("Invalid guess! Please enter a number between 1 and 10.")
        continue
    if guess == secret_number:
        print("Congratulations! You guessed the correct number.")
        break
    elif guess > secret_number:
        print("Your guess is greater than the secret number.")

    else:
        print("Your guess is lower than the secret number.")

    attempts -= 1
    print("Attempts left:", attempts)

if attempts == 0:
    print("Game Over! The secret number was:", secret_number)
```
